```python
df2= df.copy()
df2.head()
```

```python
#dropping unnecesarry columns
df2= df2.drop(['Have you received additional monetary support from your employer due to Work From Home? If yes, how much in 2020 in EUR',
             'Have you been forced to have a shorter working week (Kurzarbeit)? If yes, how many hours per week',
             'Have you lost your job due to the coronavirus outbreak?', 'Employment status', 'Annual bonus+stocks one year ago. Only answer if staying in same country',
             'Annual brutto salary (without bonus and stocks) one year ago. Only answer if staying in the same country',
             'Years of experience in Germany'], axis= 1)
df2.head()
```

```python
#renaming column headers
df2= df2.rename(columns={'Timestamp':'timestamp', 'Age':'age', 'Gender':'gender', 'City':'city', 'Position ':'position',
       'Total years of experience':'years_of_experience', 'Seniority level':'seniority_level',
       'Your main technology / programming language':'main_programming_language',
       'Other technologies/programming languages you use often':'alternative_programming_language',
       'Yearly brutto salary (without bonus and stocks) in EUR':'basic_salary',
       'Yearly bonus + stocks in EUR':'gross_salary', 'Number of vacation days':'vacation_days',
       'Сontract duration':'contract_duration', 'Main language at work':'language_at_work', 'Company size':'company_size',
       'Company type':'company_type'})
df2.head()
```
